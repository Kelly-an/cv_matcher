```python
import pandas as pd
from json import load as json_load
```

```python
filepath = '../data/2024-12-05/raw_data_0_49.json'
```

```python
with open(filepath, 'r') as f:
    data = json_load(f)
```

```python
df = pd.DataFrame(data['resultats'])
```

```python
df.head(1)
```
